```
def correct_column_names(df, expected_columns):
    current_columns = df.columns.tolist()
    corrected_columns = []
    for col in expected_columns:
        if col in current_columns:
            corrected_columns.append(col)
        else:
            # Attempt to find a close match (case insensitive)
            matches = [c for c in current_columns if c.lower() == col.lower()]
            if matches:
                corrected_columns.append(matches[0])
            else:
                # Column not found, fill with None or raise error
                corrected_columns.append(None)
    # Optionally rename columns in df to match expected_columns where possible
    rename_map = {}
    for expected, current in zip(expected_columns, corrected_columns):
        if current is not None and expected != current:
            rename_map[current] = expected
    df = df.rename(columns=rename_map)
    return df

# Example usage:
# expected_cols = ['Name', 'Age', 'Gender']
# df = correct_column_names(df, expected_cols)
# Now you can safely access df[expected_cols] without KeyError.
```
